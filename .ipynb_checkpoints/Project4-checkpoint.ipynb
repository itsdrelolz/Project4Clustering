# Clustering is an unsupervised machine learning technique that help unveil inherent structures or patterns that exist in a dataset. Clustering does this by grouping similar datapoints closer together. The goal of clustering is to make it so that the data points that exist in these clusters are more similar to the points around them than to the other clusters.
